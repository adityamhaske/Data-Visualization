```python
import numpy as np
import matplotlib.pyplot as plt
```

A relatively large number of data points to simulate a realistic dataset.



```python
np.random.seed(42)

# Number of data points
n = 1000
```

A synthetic dataset that combines two distinct distributions: a normal distribution and a heavy-tailed distribution

```python
normal_data = np.random.normal(0, 1, n)
cauchy_data = np.random.standard_cauchy(n)
weird_data = np.concatenate([normal_data, cauchy_data])
```

Making sure dataset has significant skewness, either positively or negatively, to mimic real-world scenarios where data might not follow a normal distribution

```python
weird_data = np.power(weird_data, 3)
```

```python
num_bins = 30
```

In this weird dataset, the Freedman-Diaconis Rule may break because it relies on the IQR to estimate the bin width. The presence of extreme skewness in the dataset can result in an exceptionally large IQR, leading to an overly wide bin width.

```python
plt.figure(figsize=(6, 5))
plt.hist(weird_data, bins=num_bins, color='blue', alpha=0.7)
plt.title("Histogram with Freedman-Diaconis Rule (Reduced Bins)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

The histogram created using the Freedman-Diaconis Rule in this scenario would likely have very few bins, each with a wide range.

```python
plt.figure(figsize=(6, 5))
plt.hist(weird_data, bins='scott', color='green', alpha=0.7)
plt.title("Histogram with Scott's Rule")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

To mitigate this issue, an alternative binning method that is less sensitive to extreme outliers and skewness, such as Sturges' Rule or Scott's Rule, might be more appropriate for creating a histogram
