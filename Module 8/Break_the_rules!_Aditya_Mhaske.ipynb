```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set seed for reproducibility
np.random.seed(42)
```

```python
# Generate the dataset by combining two normal distributions
data_mode1 = np.random.normal(loc=10, scale=1, size=300)  # Smaller mode
data_mode2 = np.random.normal(loc=50, scale=10, size=700)  # Larger mode
artificial_data = np.concatenate((data_mode1, data_mode2))
```

```python
# Define the range for the histogram
data_min, data_max = artificial_data.min(), artificial_data.max()

# Different bin sizes
bin_sizes = [5, 15, 30, 50]
```

```python
# Create subplots for histograms with varying bin sizes
plt.figure(figsize=(12, 6))
for i, num_bins in enumerate(bin_sizes):
    plt.subplot(2, 2, i + 1)
    plt.hist(artificial_data, bins=num_bins, alpha=0.7,)
    plt.title(f'Number of Bins: {num_bins}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
