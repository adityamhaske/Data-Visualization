Now, let's plot using the **`tophat`** and **`gaussian`** kernels using the different sample sizes we used before. We'll do this using scikit-learn's `KernelDensity()` function.

```python
from sklearn.neighbors import KernelDensity
sample_sizes = [10, 50, 100, 500, 1000, 2000]

def plot_kde(X_plot, sample_sizes, movies, kernel='gaussian'):
    for sample_size in sample_sizes:
        sample = np.array(movies['IMDB_Rating'].sample(sample_size))
        sample = sample[:, np.newaxis]
        sample = sample[~np.isnan(sample)]
        sample = sample.reshape(-1, 1)
        kde = KernelDensity(kernel=kernel).fit(sample)
        log_density = kde.score_samples(X_plot)
        plt.plot(X_plot[:, 0], np.exp(log_density), '-')

fig, ax = plt.subplots(figsize=(10,5))

# Let's look at the gaussian kernel like before
# put it in the first spot on a 1x2 grid
plt.subplot(1, 2, 1)
plt.title("Kernel: Gaussian")
plot_kde(X_plot, sample_sizes, movies)
plt.legend()
# Now let's look at the tophat kernel
# put it in the second spot on a 1x2 grid
plt.subplot(1, 2, 2)
plt.title("Kernel: Tophat")
plot_kde(X_plot, sample_sizes, movies, kernel='tophat')
plt.legend()

plt.tight_layout()
plt.show()
```

Let's try all kernel types using scikit-learn's `KernelDensity()`. Plot a 2x3 grid, however, instead of changing sample sizes like above, make each plot use the 6 different kernels supported by scikit-learn.

```python
# YOUR SOLUTION HERE
kernel_types = ['gaussian', 'tophat', 'epanechnikov', 'exponential', 'linear', 'cosine']

fig, axs = plt.subplots(2, 3, figsize=(10, 8))

for idx, kernel in enumerate(kernel_types, 1):
    plt.subplot(2, 3, idx)
    for _ in range(5):
        sample = np.array(movies['IMDB_Rating'].sample(2))
        sample = sample[~np.isnan(sample)]
        sample = sample.reshape(-1, 1)
        kde = KernelDensity(kernel=kernel).fit(sample)
        log_density = kde.score_samples(X_plot)
        plt.plot(X_plot[:, 0], np.exp(log_density), '-')
    plt.title(f"Kernel: {kernel}\nSample Size: 2")

plt.tight_layout()
plt.show()
```

```python
# YOUR SOLUTION HERE
```

The selection of bandwidth in Kernel Density Estimation (KDE) holds substantial influence over the form and smoothness of the estimated probability density function. Below, we delve into how varying bandwidth values impact the visualization:

Scott's Rule and Silverman's Rule (bw='scott' and bw='silverman'): These are automatic bandwidth selection methods that gauge bandwidth based on the distribution of the data. They tend to yield KDE plots that are notably smooth and well-balanced.

Small Bandwidths (e.g., 0.01 and 0.1): Opting for smaller bandwidths results in KDEs that are narrow and peaked. These KDEs closely track the data points, potentially leading to overfitting.

Moderate Bandwidth (bw=1): A moderate bandwidth setting produces a smoother KDE that more accurately approximates the underlying distribution. It strikes a balance between capturing local nuances and providing a holistic view of the data's shape.

Large Bandwidth (bw=5): A larger bandwidth setting generates a very smooth and broad KDE. This can oversmooth the data and obscure vital local variations or bimodal characteristics within the distribution.

In conclusion, the choice of bandwidth involves a trade-off between capturing fine-grained details in the data (small bandwidth) and creating a smooth, generalized representation (large bandwidth). The ideal bandwidth choice hinges on the specific dataset and the objectives of the analysis.

## Moving average

Pandas has a nice method called `rolling()`: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.rolling.html

It lets you do operations on the `rolling` windows. For instance, if you want to calculate the moving average, you can simply

```python
ax = co2[-100:].plot(lw=0.5)
co2[-100:].rolling(12).mean().plot(ax=ax)
```

By default, it consider every data point inside each window equally (`win_type=None`) but there are many window types supported by `scipy`. Also by default, the mean value is put at the right end of the window (trailing average).  

**Q: can you create a plot with `triang` window type and centered average?**

```python
# YOUR SOLUTION HERE
ax = co2[-100:].plot(lw=0.5)
co2[-100:].rolling(12, win_type='triang', center=True).mean().plot(ax=ax)
```
